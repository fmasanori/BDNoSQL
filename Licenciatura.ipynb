```python
import pandas as pd

%matplotlib notebook
low_memory=False
dados = pd.read_csv('MICRODADOS_ENADE_2017.txt', delimiter=";")
```

Após a leitura dos dados, é filtrado e divido a quantidade de participantes por tipo de Instituição:

1 = Pública Federal<br>
2 = Pública Estadual<br>
3 = Pública Municipal<br>
4 = Privada com fins lucrativos<br>
5 = Privada sem fins lucrativos<br>
7 = Especial<br>

```python
dados['CO_CATEGAD'].value_counts().to_frame().T.rename(columns={1:'Pública Federal', 2:'Pública Estadual', 3:'Pública Municipal',4:'Privada COM Lucrativos',5:'Privada SEM Lucrativos', 7:'Especial'}).plot(kind='bar')
dados['CO_CATEGAD'].value_counts().to_frame().T.rename(columns={1:'Pública Federal', 2:'Pública Estadual', 3:'Pública Municipal',4:'Privada COM Lucrativos',5:'Privada SEM Lucrativos', 7:'Especial'})
```
